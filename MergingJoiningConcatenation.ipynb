```python
import numpy as np
import pandas as pd
```

```python
employees = pd.DataFrame({
    'employee_id': [1, 2, 3, 4, 5],
    'name': ['John', 'Anna', 'Peter', 'Linda', 'Bob'],
    'department': ['HR', 'IT', 'Finance', 'IT', 'HR']
})

# DataFrame 2: Salary information
salaries = pd.DataFrame({
    'employee_id': [1, 2, 3, 6, 7],
    'salary': [60000, 80000, 65000, 70000, 90000],
    'bonus': [5000, 10000, 7000, 8000, 12000]
})
```

```python
employees
```

```python
salaries
```

```python
pd.merge(employees,salaries) # Default it uses common key which is employee_id
```

```python
pd.merge(employees,salaries,on="employee_id",how="inner") # This is default 
#inner mean merge and show on the base of 1,2,3 ids means inner data
```

```python
pd.merge(employees,salaries,on="employee_id",how="outer")
# Outer mean full data show even the ids are not available for match
#Actually it uses coulmn or attribues names for merging which is same because
#both have employee_id column and we get Nan where id 
#Here you can see Name and department are missing salaries coloumn
# and salary and bonus are missing in employee columns so we get Nan there
#for filling
#And Nan are dangerous
```

```python
pd.merge(employees,salaries,on="employee_id",how="left")
#here we get all the values of colums from the employee table and
# get NaN in salary and bonus table because we will all the data for left table and if there is salary and bonus values are missing 
# then we get Nan there
```

```python
pd.merge(employees,salaries,on="employee_id",how="right")
```

<h3>Concatinate The dataframe</h3>

```python
pd.concat((employees,salaries))
```

```python
pd.concat((salaries,employees),axis=1)
# axis 1 mean the columns of data table 1 and table 2 will be merged
```
