<h3>Pandas Series</h3>

```python
import numpy as np
import pandas as pd
```

```python
list=["Ahmad","20","married","Programmer"]
labels=["name","age","status","profession"]
numpyList=np.array(list)
dictionary={"name":"Ahmad","age":20,"status":"married","profession":"Programmer"}
```

```python
pd.Series(list,index=labels)
```

```python
pd.Series(numpyList,labels)
```

```python
pd.Series(dictionary)
```

So you have to come to know that we try to delete the "Designation" column but its only print the snapshot not actually delete from the RAM or disk. But we can delete it using a parameter called inplace

```python
df1.drop("Designation",axis=1,inplace=True)
```
